#### Median of median home values across all Boston Suburbs in dollars

#### Average of median home values across all Boston Suburbs in dollars 

```python
boston_data.medv.mean() * 1000
```

#### Median home value of the suburb with the newest houses

```python
newest_suburb = boston_data.loc[boston_data['age'] == boston_data.age.min()]
newest_suburb.medv.mean() * 1000
```

### Per-capita crime rate and the pupil-teacher ratio

```python
crime = boston_data.crim
teachers = boston_data.ptratio
river = boston_data.chas
color = ['red' if r == 0 else 'blue' for r in river]
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(teachers, crime, c=color, alpha=0.2)
plt.title('Per-capita crime rate and the pupil-teacher ratio\nBlue = bounded by the Charles River')
plt.xlabel("Pupil-teacher ratio [$teachers$]")
plt.ylabel("Per capita crime rates")
plt.show()
```

#### Reflections
This graph might be more valuable if outliers were removed (particularly around the pupil-teacher ratio of 20). Even with outliers removed, crime rates of neighborhoods with a 20 student per teacher rate exhibit a much larger range than neighborhoods with other pupil-teacher ratios. There may be other factors affecting the per capita crime rate.

### Proportion of black citizens and the distance to employment centers in Boston

```python
black_citizens = boston_data.black/10
distance = boston_data.dis
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(black_citizens, distance, c='green', alpha=0.2)
plt.title('Percentage of black citizens and the avg distance to employment centers\n in Boston by suburb')
plt.xlabel("% of black citizens")
plt.ylabel("Weighted avg distance to employment ctrs")
plt.show()
```

#### Reflections
This plot suggests that the majority of black citizens live in neighborhoods that are comprised of more than 35% black citizens. Neighborhoods with less than 35% black citizens tend to be 1-4 units of distance away from employment centers, while the distance from employment centers increases dramatically as the percentage of black citizens approaches 40%.

### Median value of owner-occuped homes and nitric oxide concentration along with median home value and the proportion of non-retail business

```python
x1 = boston_data.nox
x2 = boston_data.indus

y = boston_data.medv

plt.figure(figsize=(12, 6)) 
plt.subplot(1, 2, 1)
plt.scatter(x1, y, c='red', alpha=0.1)
plt.title('Med. home value vs. N2O concentration')
plt.ylabel('Median home value (in $1000s)')
plt.xlabel('Concentration of N2O')

plt.subplot(1, 2, 2)
plt.scatter(x2, y, c='blue', alpha=0.1)
plt.title('Med. home value vs. non-retail business')
plt.xlabel('Proportion of non-retail business')

plt.show()
```

#### Reflections
The subplots mapping median home values against nitric oxide concentration and the proportion of non-retail businesses look similar. There may be a correlation between the presence of nitric oxide levels and industrial activity. As home values lower, the likelihood of higher nitric oxide/higher proportion of industrial businesses nearby increases.
