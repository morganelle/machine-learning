```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
boston_data = pd.read_csv("data/boston_housing_data.csv", sep=',')
```

```python
boston_data
```

```python
len(boston_data.dropna())
```

#### Median of median home values across all Boston Suburbs in dollars 

```python
boston_data.medv.median() * 1000
```

#### Average of median home values across all Boston Suburbs in dollars 

```python
boston_data.medv.mean() * 1000
```

#### Median home value of the suburb with the newest houses

```python
newest_suburb = boston_data.loc[boston_data['age'] == boston_data.age.min()]
newest_suburb.medv.mean() * 1000
```

#### Per-capita crime rate and the pupil-teacher ratio

```python
crime = boston_data.crim
teachers = boston_data.ptratio
river = boston_data.chas
color = ['red' if r == 0 else 'blue' for r in river]
```

```python
plt.scatter(teachers, crime, c=color)
plt.title('Per-capita crime rate and the pupil-teacher ratio\nBlue = bounded by the Charles River')
plt.xlabel("Pupil-teacher ratio [$teachers$]")
plt.ylabel("Per capita crime rates [$teachers = e^crime$]")
plt.show()
```

#### Proportion of black citizens and the distance to employment centers in Boston

```python
black_citizens = boston_data.black/10
distance = boston_data.dis
```

```python
plt.scatter(black_citizens, distance)
plt.title('Percentage of black citizens and the avg distance to employment centers\n in Boston by suburb')
plt.xlabel("% of black citizens")
plt.ylabel("Weighted avg distance to employment ctrs")
plt.show()
```
